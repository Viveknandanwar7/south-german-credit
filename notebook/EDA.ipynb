```python
credit_risk_labels = {0: "Bad Credit", 1: "Good Credit"}
df['kredit'] = df['kredit'].replace(credit_risk_labels)

fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df,x='hoehe',bins=30,kde=True,color='g',ax=axs[0])
axs[0].set_title("Distribution of 'amount'")
axs[0].set_xlabel("amount")
axs[0].set_ylabel("Density")
plt.subplot(122)
sns.histplot(data=df,x='hoehe',kde=True,hue='kredit',ax=axs[1])
axs[1].set_title("Distribution of 'amount' by Credit Risk")
axs[1].set_xlabel("amount")
axs[1].set_ylabel("Density")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df, x='alter', bins=30,kde=True,color='g',ax=axs[0])
axs[0].set_title("Distribution of Age")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Density")
plt.subplot(122)
sns.histplot(data=df,x='alter',kde=True,hue='kredit',ax=axs[1])
axs[1].set_title("Distribution of 'Age' by Credit Risk")
axs[1].set_xlabel("Age")
axs[1].set_ylabel("Density")
plt.show()
```

```python
credit_counts = df['kredit'].value_counts()

credit_labels = credit_counts.index.to_numpy()  
credit_values = credit_counts.to_numpy()
plt.pie(credit_values, autopct="%1.1f%%", startangle=140)  

plt.title("Distribution of Credit Risk")
plt.legend(credit_labels, loc="best")
plt.show()
```
